```python
import pandas as pd

# Assuming you have a DataFrame named 'df' with a 'Label' column
# Replace 'your_dataset.csv' with the actual file name and path if you're loading from a CSV file
# df = pd.read_csv('your_dataset.csv')

# Replace 'DDoS' with 1 and all other values with 0 in the 'Label' column
df['Label'] = df['Label'].apply(lambda x: 1 if x == 'DDoS' else 0)

# Display the modified DataFrame
print(df)
```

```python
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt

# Assuming df is your DataFrame
X = df.drop(columns=['Label', 'Flow ID', 'Src IP', 'Dst IP', 'Timestamp'])
y = df['Label']

# Initialize the Random Forest Classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)

# Fit the model
rf_classifier.fit(X, y)

# Get feature importances
feature_importances = rf_classifier.feature_importances_

# Create a DataFrame to visualize feature importances
feature_importance_df = pd.DataFrame({'Feature': X.columns, 'Importance': feature_importances})

# Sort the DataFrame by importance
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

# Plot the feature importances
plt.figure(figsize=(14,6))
plt.barh(feature_importance_df['Feature'], feature_importance_df['Importance'])
plt.xlabel('Importance')
plt.title('Random Forest Feature Importance')
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import mutual_info_classif
import matplotlib.pyplot as plt

# Assuming df is your DataFrame
X = df.drop(columns=['Label', 'Flow ID', 'Src IP', 'Dst IP', 'Timestamp'])
y = df['Label']

# Calculate Information Gain
information_gains = mutual_info_classif(X, y)

# Create a DataFrame to visualize information gains
info_gain_df = pd.DataFrame({'Feature': X.columns, 'Information Gain': information_gains})

# Sort the DataFrame by information gain
info_gain_df = info_gain_df.sort_values(by='Information Gain', ascending=False)

# Plot the information gains
plt.figure(figsize=(20,6))
plt.barh(info_gain_df['Feature'], info_gain_df['Information Gain'])
plt.xlabel('Information Gain')
plt.title('Information Gain Feature Importance')
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import chi2
import matplotlib.pyplot as plt

# Assuming df is your DataFrame
X = df.drop(columns=['Label', 'Flow ID', 'Src IP', 'Dst IP', 'Timestamp'])
y = df['Label']

# Replace negative values with zeros
X_non_negative = X.clip(lower=0)

# Calculate Chi-square
chi2_scores, _ = chi2(X_non_negative, y)

# Create a DataFrame to visualize chi-square scores
chi2_df = pd.DataFrame({'Feature': X.columns, 'Chi-square': chi2_scores})

# Sort the DataFrame by chi-square score
chi2_df = chi2_df.sort_values(by='Chi-square', ascending=False)

# Plot the chi-square scores
plt.figure(figsize=(14,6))
plt.barh(chi2_df['Feature'], chi2_df['Chi-square'])
plt.xlabel('Chi-square')
plt.title('Chi-square Feature Importance (Ignoring Negative Values)')
plt.show()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Select relevant features
selected_features = ['Bwd IAT Min', 'Bwd IAT Mean', 'Src Port', 'Protocol', 'Bwd IAT Tot',
                     'Flow IAT Max', 'Dst Port', 'Bwd Header Len', 'Flow IAT Min', 'Flow Pkts/s',
                     'Bwd IAT Max', 'Flow Duration', 'Bwd Pkts/s', 'Flow IAT Std']

# Extract the selected features
X = df[selected_features]

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply K-means clustering
num_clusters = 8  # Number of clusters based on the labels
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)

# Define a mapping from numeric cluster labels to class labels
cluster_label_mapping = {
    0: 'U2R',  # Replace with the correct mapping for your data
    1: 'BFA',
    2: 'DDoS',
    3: 'DoS',
    4: 'Probe',
    5: 'Normal',
    6: 'Web-Attack',
    7: 'BOTNET'
}

# Replace numeric cluster labels with class labels in the dataframe
df['Cluster'] = df['Cluster'].map(cluster_label_mapping)

# Define colors for each cluster
cluster_colors = {
    'U2R': 'blue',
    'BFA': 'green',
    'DDoS': 'red',
    'DoS': 'purple',
    'Probe': 'orange',
    'Normal': 'cyan',
    'Web-Attack': 'brown',
    'BOTNET': 'pink'
}

# Visualize clusters with different colors
for cluster, color in cluster_colors.items():
    cluster_data = df[df['Cluster'] == cluster]
    plt.scatter(cluster_data['Bwd IAT Min'], cluster_data['Bwd IAT Mean'], label=f'Cluster {cluster}', color=color)

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='x', s=200, c='black', label='Centroids')
plt.xlabel('Bwd IAT Min')
plt.ylabel('Bwd IAT Mean')
plt.title('Clustering Visualization')
plt.legend()
plt.show()

# Save the results to a new CSV file
df.to_csv("clustered_data_with_labels.csv", index=False)

# Save source IPs with label 'DDoS' to a new CSV file
ddos_ips = df[df['Label'] == 'DDoS']['Src IP']
ddos_ips.to_csv("ddos_ips.csv", index=False)
```

```python
import pandas as pd
from sklearn.cluster import Birch
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Select the features for clustering
selected_features = [
    'Bwd IAT Min', 'Bwd IAT Mean', 'Src Port', 'Protocol', 'Bwd IAT Tot',
    'Flow IAT Max', 'Dst Port', 'Bwd Header Len', 'Flow IAT Min', 'Flow Pkts/s',
    'Bwd IAT Max', 'Flow Duration', 'Bwd Pkts/s', 'Flow IAT Std'
]

# Extract the selected features from the DataFrame
X = df[selected_features]

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply Birch clustering
birch = Birch(n_clusters=3, threshold=0.5, branching_factor=50)
df['Cluster'] = birch.fit_predict(X_scaled)

# Visualize the clusters (you can modify this based on the number of features)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=df['Cluster'], cmap='viridis', s=50, alpha=0.5)
plt.xlabel('Bwd IAT Min (Standardized)')
plt.ylabel('Bwd IAT Mean (Standardized)')
plt.title('Birch Clustering of Network Traffic')
plt.show()

# Analyze the clusters and identify patterns
# Note: Birch does have subcluster centers, so you can analyze those as well
subcluster_centers = scaler.inverse_transform(birch.subcluster_centers_)
subcluster_centers_df = pd.DataFrame(subcluster_centers, columns=selected_features)

# Display subcluster centers (representative values for each subcluster)
print("Subcluster Centers:")
print(subcluster_centers_df)
```

```python
from sklearn.metrics.cluster import adjusted_rand_score

# Drop rows with missing values in either 'Label' or 'Cluster' columns
df_cleaned = df.dropna(subset=['Label', 'Cluster'])

# Assuming 'Label' is your ground truth
true_labels = df_cleaned['Label'].map({'normal': 0, 'ddos': 1}).values

# Drop rows with missing values in the 'Cluster' column
df_cleaned = df_cleaned.dropna(subset=['Cluster'])

# Ensure the 'Cluster' column has integer values
df_cleaned['Cluster'] = df_cleaned['Cluster'].astype(float)  # Convert to float to handle NaN
df_cleaned['Cluster'] = df_cleaned['Cluster'].fillna(0)      # Replace NaN with 0
cluster_labels = df_cleaned['Cluster'].astype(int).values

# Check for NaN values in true_labels and cluster_labels
nan_indices_true_labels = pd.isnull(true_labels)
nan_indices_cluster_labels = pd.isnull(cluster_labels)

# Remove NaN values from true_labels and cluster_labels
true_labels = true_labels[~nan_indices_true_labels]
cluster_labels = cluster_labels[~nan_indices_cluster_labels]

# Convert to integer type
true_labels = true_labels.astype(int)
cluster_labels = cluster_labels.astype(int)

# Ensure consistent lengths
min_length = min(len(true_labels), len(cluster_labels))
true_labels = true_labels[:min_length]
cluster_labels = cluster_labels[:min_length]

# Calculate ARI
ari_score = adjusted_rand_score(true_labels, cluster_labels)

print("Adjusted Rand Index (ARI):", ari_score)
```

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score
from sklearn.cluster import Birch, KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Select the features for clustering
selected_features = [
    'Bwd IAT Min', 'Bwd IAT Mean', 'Src Port', 'Protocol', 'Bwd IAT Tot',
    'Flow IAT Max', 'Dst Port', 'Bwd Header Len', 'Flow IAT Min', 'Flow Pkts/s',
    'Bwd IAT Max', 'Flow Duration', 'Bwd Pkts/s', 'Flow IAT Std'
]

# Extract the selected features from the DataFrame
X = df[selected_features]

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Choose the number of clusters for evaluation
num_clusters = 2

# Birch clustering
birch = Birch(n_clusters=num_clusters, threshold=0.5, branching_factor=50)
df['Cluster_Birch'] = birch.fit_predict(X_scaled)

# K-Means clustering
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
df['Cluster_KMeans'] = kmeans.fit_predict(X_scaled)

# Evaluate clustering using silhouette score
silhouette_score_birch = silhouette_score(X_scaled, df['Cluster_Birch'])
silhouette_score_kmeans = silhouette_score(X_scaled, df['Cluster_KMeans'])

print(f"Silhouette Score (Birch): {silhouette_score_birch:.3f}")
print(f"Silhouette Score (K-Means): {silhouette_score_kmeans:.3f}")

# Evaluate clustering using Davies-Bouldin index
davies_bouldin_score_birch = davies_bouldin_score(X_scaled, df['Cluster_Birch'])
davies_bouldin_score_kmeans = davies_bouldin_score(X_scaled, df['Cluster_KMeans'])

print(f"Davies-Bouldin Index (Birch): {davies_bouldin_score_birch:.3f}")
print(f"Davies-Bouldin Index (K-Means): {davies_bouldin_score_kmeans:.3f}")
```
